### **ENCODING THE DATA USING LABEL ENCODER.**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df["neighbourhood"] = le.fit_transform(df["neighbourhood"])
df["room_type"] = le.fit_transform(df["room_type"])
df.head()
```

###**SPLITTING THE DATA IN DEPENDENT & INDEPENDENT SET.**

```python
x = df.iloc[: , :-1].values
y = df.iloc[: , -1].values
```

```python
x
```

```python
y
```

### **SCALING THE DATA ON SAME UNIT.**

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x = sc.fit_transform(x)
```

```python
x
```

### **SPLITTING THE DATA IN TRAINING AND TESTING SETS**

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y , test_size = 0.33 , random_state = 1)
```

```python
xtrain
```

```python
ytrain
```

#**MODEL CREATION**

```python
import tensorflow as tf

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
#Initialising the network
ann = Sequential()

#Adding layers to the network
ann.add(Dense(units=10, activation="relu")) #1st input layer
ann.add(Dense(units=5, activation="relu")) #2nd input layer
ann.add(Dense(units=4, activation="relu")) #3rd input layer
ann.add(Dense(units=2, activation="relu")) #4th input layer

#Creating the output layer
ann.add(Dense(units=1))

#Establish the connection between the layers
ann.compile(optimizer="adam", loss="mse")

#Training the model
ann.fit(xtrain,ytrain , epochs = 800 , validation_data= (xtest,ytest))
```

**Plotting the loss graph**

```python
loss = pd.DataFrame(ann.history.history)
loss.plot()
```

```python
ann.evaluate(xtrain,ytrain)
```

```python
ann.evaluate(xtest,ytest)
```

**IMPROVING ACCURACY**

***Early Stop Method***

```python
#Initialising the network
ann = Sequential()

#Adding layers to the network
ann.add(Dense(units=10, activation="relu")) #1st input layer
ann.add(Dense(units=5, activation="relu")) #2nd input layer
ann.add(Dense(units=4, activation="relu")) #3rd input layer
ann.add(Dense(units=2, activation="relu")) #4th input layer

#Creating the output layer
ann.add(Dense(units=1))

#Establish the connection between the layers
ann.compile(optimizer="adam", loss="mse")

from tensorflow.keras.callbacks import EarlyStopping
earlystop = EarlyStopping(monitor = "val_loss",mode = "min", verbose = 1 , patience = 20)

ann.fit(xtrain,ytrain , epochs = 800 , validation_data= (xtest,ytest) , callbacks=[earlystop])
```

**Plotting the loss graph**

```python
loss = pd.DataFrame(ann.history.history)
loss.plot()
```

```python
ann.evaluate(xtrain,ytrain)
```

```python
ann.evaluate(xtest,ytest)
```

**Showcasing the prediction.**

```python
ypred = ann.predict(xtest)
```

```python
dic = {"Actual Y":ytest , "Predicted Y":ypred.flatten()}

df1 = pd.DataFrame(dic)
```

```python
df1.tail()
```

```python
from sklearn.metrics import mean_squared_error

print(f"MSE :- {mean_squared_error(ytest,ypred)}")
print(f"RMSE :- {np.sqrt(mean_squared_error(ytest,ypred))}")
```
